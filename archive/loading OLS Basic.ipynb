```python
import pandas as pd
import numpy as np
import pickle
```

```python
nations = ["US", "CH", "JPN", "UK"]

tables = []

for nation in nations:
    filename = "./data/" + nation + 'data.pickle' 
    with open(filename, 'rb') as handle:
        df = pickle.load(handle)
        tables.append(df)
```

```python
us, ch, jpn, uk = tables 
```

```python
def descriptivestats(df,statistics): #enter a dataframe and additional statistics you like to include
    desc = df.describe()
    return desc.append(df.reindex_axis(desc.columns,1).agg(statistics))

np.round(descriptivestats(us,['kurt','skew']),2).T

# Keine Ahnung wie ich mehrere Grafiken ausgeben lassen kann in einem Feld sonst halt über mehrere Felder verteilen

np.round(descriptivestats(ch,['kurt','skew']),2).T
np.round(descriptivestats(jpn,['kurt','skew']),2).T
np.round(descriptivestats(uk,['kurt','skew']),2).T
```
